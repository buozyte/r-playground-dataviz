## Grammar of Graphics

Visualizaion theory with 3 key priciples (e.g. implemented in `ggplot2`!):
1. separate data from aestetics
2. define common plot/chrt elements
3. composition of common elements

```r
# example where this grammar is used
library(gapminder)
gm_dt <- as.data.table(gapminder)[year %in% c(1977, 2007)]

ggplot(data=gm_dt, aes(x=gdpPercap, y=lifeExp)) +                                 # data and aestatics separated
  geom_point(aes(color=continent, size=pop)) +
  facet_grid(~year) + scale_x_log10() +
  labs(y="Life expectancy at birth", x="per-capita GDP", size = 'Population')
```

## Components of the Layered Grammer

### 1. Components of the Grammar of Graphics

Mayor layers (used to combine to plots):
* always used:
    * Data: object, where columns = values (e.g. `data.table`, `data.frame`) (~`data`)
    * Aestetics: map data to visual characteristics (e.g. position (x,y), color, size, shape,) (~`aes`)
    * Layers/geometric objects: geomatric representation defining type of plot data (e.g. points, lines, boxplots) (~`geom_{}`)
* often used:
    * Scales: how to convert visual characteristic to display values (for each aestetic) (e.g. log, color, size, shape scales) (~`scale_{}`)
    * Facets: how to split data in subsets + display in multiple subgraphs (~`facet_{}`)
* useful, but with care:
    * Stats: statistical trafos (typically summarize data) (e.g. counts, means, medians, regression lines) (~`stat`)
* domain-specific usage:
    * Coordinate system: 2D space that data is projected on (e.g. cartesian, polar, map coordinates) (~`coord_{}`)

### 3. Mapping of Aestetics

#### 3.1 Mapping of Color, Shape and Size

The `color`/`colour`, `shape` and `size` refine the plot to make certain information visible or better visible.

Hint: can all be used in one `aes()`!

```r
ggplot(data = gm_dt, aes(x=gdpPercap, y=lifeExp, color=continent)) + geom_point()
ggplot(data = gm_dt, aes(x=gdpPercap, y=lifeExp, shape=continent)) + geom_point()
ggplot(data = gm_dt, aes(x=gdpPercap, y=lifeExp, color=continent, size=pop)) + geom_point()
```

#### 3.2 Global vs. Individual Mapping

`aes` can be defined either for one plot (e.g. inside `geom_{}()`) or for all following plots (e.g. inside `ggplot()`) (globally).

Hint: global and "local" can be combined. The color can for example be set globally, while the size can depend on the plot.

```r
# the following example result in the same plot
ggplot(data = gm_dt) + 
  geom_point(aes(x=gdpPercap, y=lifeExp)) +
  geom_smooth(aes(x=gdpPercap, y=lifeExp))              # "local" definition

ggplot(data = gm_dt, aes(x=gdpPercap, y=lifeExp)) +     # global definion
  geom_point() +
  geom_smooth()                                         # equivalent to `stat_smooth()` -> statistical trafos
                                                        #   summarizing data (i.e. counts, means, medians, ...)
```

### 4. Facets, Axes and Labels

Facets: split data and create subplots
Labels: add via `labs()`, can be used for axes, legend, ...

```r
# example: add new lables for the axes and the legend
ggplot(data=gm_dt, aes(x=gdpPercap, y=lifeExp)) + geom_point(aes(color=continent, size=pop)) +
  facet_grid(~year) + scale_x_log10() +
  labs(x="per-capita GDP", y="Life expectancy at birth", size = 'Population')
```

## Different Types of One- and Two-Dimensional Plots

### 1. Plots for One Single Continuous Variable

#### 1.1 Histograms (`geom_histogram()`)

Represent freuqencies of values in a variable (bucketed into ranges). Similar to bar plot, but groups values into continuous ranges.

Use `bin` or `binwidth` argument to choose a fitting amount of bins or fitting bin sizes.

Hint: input must be numeric variables!

```r
data=data.frame(value=rnorm(100))
ggplot(data, aes(x=value)) +
  geom_histogram()
  
ggplot(data, aes(x=value)) +
  geom_histogram(bins=10)
```

#### 1.2 Density Plots (`geom_density()`)

Variation of histograms used to represent distribution of numeric variables. Distribution usually obtained by kernel density estimation to smoothen out the noise. ("Smooth across bina and not affected by number of bins".)

Use `bw` argument to set the bandwidth of the plot manually (have huge impact on the plot, be careful!).

Hint: input must be numeric variables!

```r
data=data.frame(value=rnorm(100))
ggplot(data, aes(x=value)) +
  geom_density()
  
ggplot(data, aes(x=value)) +
  geom_density(bw=0.1)
  
ggplot(data, aes(x=value)) +
  geom_density(bw=1.1)
```

#### 1.3 Boxplots (`geom_boxplot()`)

Graphical insight into distribution, as they show median, quartiles and extreme values.

Based on four values:
1. median: center of data, i.e. middle value of sorted list ("second quartile")
2. first quartile: 25% quartile of the data
3. third quartile: 75% quartile of the data
4. interquartile range: distance between first and third quartile

Especially useful for non-Gaussian (non-)symmertic data and exponentially distributed data. Not suitable for bimodal distribution as they only show one mode (i.e. the median) or for discrete data.

---

Box: interquartile range

Line, horizontal (usually in box): Median

Lines, vertical: first and last 25% of the data (outside of interquartile range), i.e. lines to minimal and maximal value

Scatter above and below lines: potential outliers

(Visualization and more examples: https://www.r-graph-gallery.com/boxplot.html)

```r
ggplot(mtcars, aes(x=as.factor(cyl), y=mpg)) + 
    geom_boxplot(fill="slateblue", alpha=0.2) + 
    xlab("cyl")
```

### 2. Plots for Two Variables: One Continuous, One Discrete

#### 2.1 Barplots (`geom_bar()`)

Useful to highlight individual quantitive values per category. Can combine two attributes of 2D location and line length to encode quantative values, can therefore focus on individual values and support comparison.

```r
ggplot(data=gm_dt, aes(x=continent, y=gdpPercap)) +
  geom_bar(stat='identity', width=.7)
```

#### 2.2 Barplots with Errorbars (`geom_bar()` + `geom_errorbar()`)

Visualize uncertainties in bar plots. Consider for example standard deviation (variation of quantity in sample) or standard error of mean (how well mean is estimated).

Hint: *SEM* = *SD*/*$\sqrt{n}$* (*n* = sample size)

```r
# example: plot average `hwy` per `class` including
#           error bars computed as the average plus/minus standard deviation of `hwy`
as.data.table(mpg) %>% 
  .[, .(mean = mean(hwy),
        sd = sd(hwy)),
    by = class] %>%                                          # compute mean and standard deviation by class
  ggplot(aes(class, mean, ymax=mean+sd, ymin=mean-sd)) + 
  geom_bar(stat='identity') +                                # bar plot with mean per class
  geom_errorbar(width = 0.3)                                 # add errorbars going from ymin to ymax
```

#### 2.3 Boxplots per Category (`geom_boxplot()` + `geom_jitter()`/`geom_dotplot()`)

Use boxplot to show distributions of continuous variables w.r.t. categories, useful to compare different distributions of each category.

```r
# example: visualize `hwy` per `class`
ggplot(mpg, aes(class, hwy)) +
  geom_boxplot()
  
ggplot(mpg, aes(class, hwy)) +
  geom_boxplot() +
  geom_jitter()

ggplot(mpg, aes(class, hwy)) +
  geom_boxplot() +
  geom_dotplot(binaxis='y', stackdir='center', dotsize=0.3)
```

#### 2.4 Violin Plots (`geom_violin()`)

Alternative to boxplot.

Advantage: shows entire distribution of data, therefore good for multimodal data.

```r
ggplot(mpg, aes(class, hwy)) +
  geom_violin()
```

#### 2.5 Beanplots (`geom_beeswarm()`)

Alternative to boxplot.

Advantage: individual observations visible as small line in one-dim. scatter plot and estimated density of distributions visible.

Problem: only useful up to certain number of data points.

```r
library(ggbeeswarm)
ggplot(mpg, aes(class, hwy)) + 
  geom_beeswarm()
```

### 3. Plots for Two Continuous Variables

#### 3.1 Scatter Plots (`geom_point()`)

Easy visualization of relationship between two continuous variables.

Dots = pairs of values corresponding to the two variables.

But: problems when dealing with huge number of points.

```r
ggplot(mpg, aes(displ, hwy, color=class)) + geom_point()

ggplot(mpg, aes(displ, hwy)) + geom_point() + facet_wrap(~class)     # instead of colors: different subplots
```

##### 3.1.1 Text Labeling (`geom_text()`)

Label individual points in scatter plot.

Often helpful: `geom_text_repel()` (from `ggrepel` package) to avoid overlapping of labels.

```r
# example: without and with `repel`
library(ggrepel)
set.seed(12)
mpg_subset <- mpg[sample(1:nrow(mpg), 30, replace=FALSE),] 

ggplot(mpg_subset, aes(displ, hwy, label=manufacturer)) +
  geom_point() +
  geom_text()

ggplot(mpg_subset, aes(displ, hwy, label=manufacturer)) + 
  geom_point() +
  geom_text_repel()
```

##### 3.1.2 Log Scaling (`scale_{}()`)

Helpful if for example a few points points are notably larger than most of the other points.

Instead of scaling the points manually, use a `scale_{}` function.

```r
library(MASS)                            # for data set
animals_dt <- as.data.table(Animals)

ggplot(animals_dt, aes(x = body, y = brain)) + geom_point()

ggplot(animals_dt, aes(x = body, y = brain)) + geom_point() +
  scale_x_log10() + scale_y_log10()

ggplot(animals_dt, aes(x = body, y = brain)) + geom_point() +
  scale_x_log10() + scale_y_log10() +
  annotation_logticks()                                           # add ticks in log scale
```

#### 3.2 Density Plots in 2D (`geom_hex()`)

Useful when number of points is too huge for scatter plot, as it counts number of observations within particular area in 2D space.

```r
x <- rnorm(10000); y=x+rnorm(10000)
data.table(x, y) %>% ggplot(aes(x, y)) +
  geom_hex()
```

#### 3.3 Line Plots (`geom_line()`)

Visualize connection of series of individual data points, display trend of series of data points or show strength of movement of values through time. Useful to show shape of date as it flows/changes from point to point.

```r
ggplot(economics, aes(date, unemploy / pop)) +
  geom_line()
```

### 1. Plot Matrix

Useful for exploring distributions and correlations of few variable in matrix-like representation, i.e. create scatter plot (lower triangle) and (Pearson) correlation coefficients (upper triangle) for each pair of considered variables and a density plot with respective distribution for every single variable (diagonal).

Graphical representation of correlation matrix, useful to highlight most correlated variables in data set, e.g. colored according to value or ordered according to degree of association.

Often called: "corrgrams" or "correlograms".

```r
ggcorr(mtcars, geom = 'circle')
```
