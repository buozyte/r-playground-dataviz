## Statistical hypothesis testing

Method to assess wether observed observed trend could have arisen by chance.

### 1. Permutation testing: An intuitive build-up

Consider: possibility that large difference of growth rate medians could arise by chance when arbitrary groups of same size as those defined by genotype are used. (TODO: wtf does that mean?)

Simulation of random/permuted data: `sample()`.

```r
LETTERS[1:8]
sample(LETTERS[1:8])
```

```r
dt_permuted <- copy(dt)
set.seed(0)                                   # the seed of the random number generator
dt_permuted[, genotype:=sample(genotype)]

# %+% updates dataset of ggplot object
p <- p %+% dt_permuted 
p
```

```r
# input of function: table
# output of function: difference of median growth rates
diff_median <- function(tab){
  tab[genotype == 'Wild isolate', median(growth_rate, na.rm=T)] - 
    tab[genotype == 'Lab strain', median(growth_rate, na.rm=T)]
}

T_obs <- diff_median(dt)
T_obs

diff_median(dt_permuted)

# difference of medians is smaller!
```

```r
# number of permutations
m <- 1000 

# initialize T_permuted with missing values
# (safer than with 0's)
T_permuted <- rep(NA, m)

# iterate for i=1 to m
for(i in 1:m){
  # permute the genotype column in place 
  dt_permuted[ , genotype:=sample(genotype)]
  # store the difference of medians in the i-th entry of T_permuted
  T_permuted[i] <- diff_median(dt_permuted)
}

ggplot(data.table(T_permuted), aes(x = T_permuted)) + 
  geom_histogram() + 
  geom_vline(aes(xintercept=T_obs, color = "T_obs"))      # add original observation to permutations
```

*Conclusion*:

Repeat randomization many times to get distribution of permutations.

Observed difference significantly larger than difference of any permutation, hence unlikely to arise by chance.

### 2. Concepts of Statistical Hypothesis Testing

Overview: collected data is particular realization of a certain random process. From this data test statistic is computed. Now we asumme that underlying random process conforms to certain hypothesis ("null hypothesis"). Based on this assumption, different realization datasets can arise, for which test statistics would get different values. Then compute how likely it is to see test statistics as extreme as those from actual data. Null hypothesis is then rejected or accepted.

#### 2.1 Test Statistic

Single number summarizing data and capturing trend (more extreme statistic means stronger trend).

Here: $T = \mathrm{median}_{i \in \mathrm{Wild}}(y_{i}) - \mathrm{median}_{i \in \mathrm{Lab}}(y_{i})$.

(Could also use: difference of means, difference of means divided by within-group standard deviation, etc.. If test statistic based on random variables, then discriminating the observed data from permutation is more difficult.)

#### 2.2 Null Hypothesis

Negative control, i.e. assumption (based on the problem) that observed trend is *not* real and arose purely by chance. ("Devil's Advocate")

#### 2.3 P-Value

Distribution of test statistic under null hypothesis: $p(T|H_{0})$.

P-value: probability of obtained test statistic being same of more extreme than observed one unter the assumption that null hypothesis is true.

---

Definition of P-value:
- right-tail events: $P = p(T \geq T_{\mathrm{obs}} | H_{0})$
- left-tail events: $P = p(T \leq T_{\mathrm{obs}} | H_{0})$
- double tail events: $P = 2 \cdot \mathrm{min}\{p(T \geq T_{\mathrm{obs}} | H_{0}), p(T \leq T_{\mathrm{obs}} | H_{0})\}$ (Note: `min` is for selecting the correct tail.)

When to apply which test:
- one-tailed test: take account for one possibility (e.g. test if true difference is 0 vs. true difference is greater than 0)
- two tailed test: take account for two possibilities (e.g. test if true difference is 0 vs. true difference is not 0)

Reject null hypothesis for sufficiently small P-values, i.e. when result is statistically significant. Common practice: $P < \alpha = 0.05$.

#### 2.4 Some Intuition on Hypothesis Testing and P-Value

Scenario: legal trial for a murder, but person is innocent until proven guilty.

Evidence: murder weapon with finger prints, witness, bought chemicals used in murder.

Null hypothesis: person is innocent.

Conclusion: a lot of unlikely coincidences required for person to be innocent, hence P-value is low. Thus null hyposesis is rejected.

---

Scenario: legal trial for a murder, but person is innocent until proven guilty.

Evidence: possible witness, with bad sight.

Null hypothesis: person is innocent.

Conclusion: high P-value, thus null hypothesis is not rejected.

#### 2.5 What the P-Value is not

Not probability of observed test statistic given that the null hypothesis is true, i.e. $p(T \geq T_{\mathrm{obs}} | H_{0}) \neq p(T = T_{\mathrm{obs}} | H_{0})$.

*Example: observing a certain continuous value will be zero, thus P-values would not be useful.*

---

Not probability that null hypothesis is tru given the data, i.e. $p(T \geq T_{\mathrm{obs}} | H_{0}) \neq p(H_{0} | T = T_{\mathrm{obs}})$.

*Example: person with bad sight witnesses murder and is the only evidence. Can't convict anyone based on weak evidance. But also no evidence that person is innocent exists, i.e. "absence of evidence is not evidence of absence".*

### 3. Permutation testing, formally

Used to test statistical dependence between two variables. Test statistics can be any measure capturing dependence.

---

Assumption: observations i.i.d., i.e. data generation process same for all observations and all observations independent.

But: assumption is tricky, if for example data is longitudal or hidden groups exist. Therefore question this assumption and adress it by e.g. stratifying data.

---

Null hypothesis of permutation test: two variables are statistically independent.

Hence: under null hypothesis, data generation of these variables is independent. Combined with i.i.d. assumption, implication is that values of one variable could have occured in any other order with same likelyhood.

---

Exact permutation test: consider all possible distinct permutations. If number of observations high, drawing enough random permutations results in decent idea of distribution of $p(T|H_{0})$.

For one-sided P-value: $\hat{P} = \frac{r+1}{m+1}$ with $m=$ number of permutations and $r= \#\{T^{*} \geq T_{\mathrm{obs}}\}$.

## Confidence Intervals: Quantifying Uncertainty in Parameter Estimates

Hypothesis testing: effective way to not be fooled by random patterns, but only answers specific question.

But: often want to estimate certain quantity and report uncertainty of this estimate ("parameter estimation"). Visualization in original script.

In yeast example: concluded that true difference in growth rate is unlikely zero. But is estimated growth rate of 2.2 a good estimate of true difference?

### 1. Repeating Experiments to Quantify Uncertainty

*Intuition*:

Unlimited budget and can repeat yeast experiment 100 times with same experimental protocol. Difference of median growth rates is computed every time, which results in distribution of parameter estimates (corresponds to null distribution if null hypothesis true).

Assume: estimates quiet close to measured one and range of results is wide.

But: true difference is unkown and any estimate could be correct.

Hence: quantify uncertainty by reporting full range of estimates, even if interval is very big.

More robust alternative: report interval covering central $x$% of the values. Usually, true difference will be in this interval.

### 2. Simulating Repeated Experiments

Problem: often cannot rerun experiment due to time consumption.

Solution: simulate reruns by sampling data.

#### 2.1 Empirical distribution

Random variable X and random sample of n independent realizations drawn from it. Then empirical distribution is distribution giving equal probability to each observation.

One draw from empirical distribution = pick data point with probability $\frac{1}{n}$. Hence random draws of any size equivalent to sampling with replacement.

---

Fundamental result: empirical distribution converges to underlying distribution.

Implication: drawing from empirical distribution is justified proxy for drwaing from actual underlying distribution.

Empirical cumulative distribution function: $F(x) = \frac{1}{n} \sum_{i=1}^{n} \mathbb{I}_{x_{i} \leq x}$. Returns proportion of observation less than or equal to $x$. Equivalently, eCDF returns probability of one randomly picked abservation among the sample to be less than or equal to $x$, i,e, eCDF is commulative distribution of process of randomly drawing from sample.

```r
set.seed(10)
x <- rnorm(20)                            # 20 random numbers normally distributed 
xrnd <- sample(x, 15, replace = TRUE)     # 15 random draws from the data in x 
xrnd

plot(ecdf(x))
# interpretation of plot: about half of the observations are less than or equal to 0.
```

```r
x_small <- rnorm(5)
x_middle <- rnorm(20)
x_big <- rnorm(200)
x_lbl <- c(rep("small_n", 5), rep("medium_n", 20), rep("large_n", 200))
x_combined <- c(x_small, x_middle, x_big)
dt_ecdf <- data.table(sample_size = x_lbl, x = x_combined)
ggplot(dt_ecdf) +
  stat_ecdf(aes(x, colour = sample_size)) +
  stat_function(fun = pnorm) + 
  ylab("Fn(x)")
```

#### 2.2 Case Resampling Bootstrap

Idea: simulate experiments using the idea of drawing from empirical distribution (instead of true distribution).

Concretely: take sample of size n with replacement from observed data to produce new dataset ("case resampling bootstrap").

---

Note: new data will resemble old data, but will certainly not be the same due to sampling with replacement.

```r
# an example:
dt_resampled <- dt[sample(nrow(dt), replace = TRUE)]
diff_median(dt_resampled)
```

### 3. Quantifying Uncertainty Using the Case Resampling Bootstrap

Initial state: do $R$ radom simulations by case resampling, which returns random value for each parameter $T^{*}_{1} \leq \ldots \leq T^{*}_{R}$ (ordered).

---

Obtain confidence interval by using quantiles.

Concretely: try to cover central $(1-\alpha) \cdot 100\%$ of distribution of estimates, which is achieved by using the interval $(T^{*}_{(R+1) \cdot \alpha/2}, T^{*}_{(R+1) \cdot (1-\alpha)/2)})$

```r
# number of random simulations
R <- 1000 

# initialize T_boot with missing values
# (safer than with 0's)
T_bootstrap <- rep(NA, 1000)

# iterate for i=1 to R
for(i in 1:R){
  # sample the original data with same size with replacement
  dt_boot <- dt[sample(nrow(dt), replace=TRUE)]
  # store the difference of medians in the i-th entry of T_permuted
  T_bootstrap[i] <- diff_median(dt_boot)
}

conf_int <- quantile(T_bootstrap, c(0.025, 0.975))
conf_int
```

```r
# visualize entire distribution + observed value and 95% boostrap percentile confidence interval
ggplot(data.table(T_bootstrap), aes(T_bootstrap)) +
  geom_histogram(bins=30) + 
  geom_vline(aes(xintercept=T_obs, color = "T_obs")) +
  geom_vline(aes(xintercept=conf_int[1], color="CI"), linetype="dashed") +
  geom_vline(aes(xintercept=conf_int[2], color="CI"), linetype="dashed") +
  xlab("estimated difference of median growth rates")
```

### 4. Confidence Intervals: Formal Definition

**Definition**:

Confidence interval of confidence level $1-\alpha$ for parameter $\theta$ is an interval $C=(a,b)$, which would contain $\theta$ with probability $1-\alpha$ if data generation process would be repreated, i.e. $p(\theta \in C) \approx 1-\alpha$. Typical value: $\alpha = 0.05$.

(Note: $a, b$ are functions of data, hence $C$ is a random variable.)

---

To get intuition: instead of computing one $1-\alpha$ interval from all yeast experiments, compute separate $1-\alpha$ intervals for each experiment. Thus true difference of medians will be contained in $1-\alpha$ of the computed intervals, thus only in $\alpha$ of the cases the interval does not include true difference of medians.

### 5. Visualizing the formal definition of Confidence Intervals

Consider example where true value of parameter is known. Specifically, try to use sample mean as estimate of true mean of standard normal distribution (i.e. true mean is 0).

---

Example: draw 30 values fromstandard normal distribution, then compute sample mean and do boostrap resampling, which creates distribution of estimates. Lastly, build $95\%$ case resampling boostrap confidence interval.

Conclusion: interval covers all (also true value!) but most extreme estimates. True value is slightly lower than observed value.

```r
set.seed(100)
exp_1 <- rnorm(30)      # original data

# sample mean
observed_mean <- mean(exp_1)

# bootstrap resampling + sample mean for each simulation
boot <- lapply(1:1000, function(i){sample(exp_1, 30, replace = TRUE)})
sample_means <- sapply(boot, mean)

# 95% C.I. (given by 2.5% and 97.5% quantile)
conf_int = quantile(sample_means, c(0.025, 0.975))

bootstrap_tbl = data.table(means = sample_means)
ggplot(data = bootstrap_tbl, aes(x = means)) + 
  geom_histogram() + 
  geom_vline(aes(xintercept=observed_mean, color="observed"))  + 
  geom_vline(aes(xintercept=0, color="true"))  + 
  geom_vline(aes(xintercept=conf_int[1], color="CI"), linetype="dashed") +
  geom_vline(aes(xintercept=conf_int[2], color="CI"), linetype="dashed") +
  scale_color_manual(name = "Legend", values = c(true="black", observed = "blue", CI = "red"))
```

```r
rerun_experiment <- function(j) {
  exp <- rnorm(30)
  boot <- lapply(1:1000, function(i){sample(exp, 30, replace = TRUE)})
  sample_means <- sapply(boot, mean)
  conf_int = quantile(sample_means, c(0.025, 0.975))
  return(conf_int)
}

rerun <- sapply(1:100, rerun_experiment)

intervals <- data.table(t(rerun))
intervals[, idx := 1:100]
intervals[, mid := (`97.5%` + `2.5%`)/2]
intervals[, contains_true := ((`97.5%` >= 0) & (`2.5%` <= 0))]
ggplot(data = intervals, aes(mid, idx, color=contains_true)) + 
  geom_errorbar(aes(xmin=`2.5%`, xmax=`97.5%`)) +
  geom_vline(aes(xintercept=0)) +
  xlab("estimated mean") +
  ylab("experiment number") + 
  labs(color="Error bar contains\ntrue mean")
```

Last plot visualizes test, if interval captures true value about $95\%$ of the time, by showing confidence interval for each repetition of the experiment. Here the true value is indeed captured by intervals most of the time.

### 6. Hypothesis Testing with the Confidence Interval

Use null hypothesis as the parameter to be tested and reject it if confidence interval does not include it, i.e. "reject null hypothesis at significance level of $\alpha$".

But: rejection is based on the method, so be careful.

---

If e.g. multiple groups are compared: construct multiple confidence intervals and reject null hypothesis iff confidence intervals do not overlap.
